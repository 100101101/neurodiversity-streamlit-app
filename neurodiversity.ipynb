```python
import pandas as pd
import numpy as np

test = pd.read_excel('E://FYP//AVSRT_Autism_figs_source_data.xlsx')
train = pd.read_excel('E://FYP//AVSRT_Autism_subject_info.xlsx')

print(train.head())
print(test.head())
```

```python
print(train.info())
print(test.info())
```

```python
print(train.describe(include='all'))
print(test.describe(include='all'))
```

```python
train['VIQ'].fillna(train['VIQ'].median(), inplace=True)
train.head()
```

```python
# train = train.drop(columns=['Unnamed: 8'])
train.head()
```

```python
train['Age'].fillna(train['Age'].median(), inplace=True)
train.head()
```

```python
train['Sex'].fillna(train['Sex'].mode()[0], inplace=True)
train.head()
```

```python
print(train[['Age', 'Sex']].isnull().sum())
```

```python
print(train.columns)
```

```python
train['PIQ'].fillna(train['PIQ'].median(), inplace=True)
train['FSIQ'].fillna(train['FSIQ'].median(), inplace=True)

print(train[['PIQ', 'FSIQ', 'ADOS']].isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


plt.figure(figsize=(8,5))
sns.histplot(data=train, x='Age', kde=True)
plt.title('Distribution of Age')

plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=train, x='PIQ',kde=True, bins=15)

plt.title('Distribution of Performance IQ (PIQ)')
plt.xlabel('PIQ Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=train, x='VIQ',kde=True, bins=15)

plt.title('Distribution of Visual IQ (VIQ)')
plt.xlabel('VIQ Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=train, x='FSIQ',kde=True, bins=15)

plt.title('Distribution of Full_Scale IQ (FSIQ)')
plt.xlabel('FSIQ Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Group', data=train)
plt.title('Distribution of Groups')
plt.xlabel('Group (1: NT, 2: ASD ?)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=train)
plt.title('Distribution of Sex')
plt.xlabel('Sex (1: Male, 2: Female ?)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Sex 1', 'Sex 2']) # Optional: Label ticks
plt.show()
```

```python
# Compare Age between Groups
plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Age', data=train)
plt.title('Age Distribution by Group')
plt.xlabel('Group')
plt.ylabel('Age (years)')
plt.xticks([0, 1], ['Group 1', 'Group 2']) # Optional: Label ticks
plt.show()
```

```python
# Compare IQ Scores between Groups (Example for PIQ, repeat for VIQ, FSIQ)
plt.figure(figsize=(7, 5))
sns.boxplot(x='Group', y='PIQ', data=train)
plt.title('PIQ Distribution by Group')
plt.xlabel('Group')
plt.ylabel('PIQ Score')
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Group', y='VIQ', data=train)
plt.title('VIQ Distribution by Group')
plt.xlabel('Group')
plt.ylabel('VIQ Score')
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Group', y='FSIQ', data=train)
plt.title('FSIQ Distribution by Group')
plt.xlabel('Group')
plt.ylabel('FSIQ Score')
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.show()
```

```python
# Compare Sex distribution across Groups
plt.figure(figsize=(7, 5))
sns.countplot(x='Group', hue='Sex', data=train)
plt.title('Sex Distribution within each Group')
plt.xlabel('Group')
plt.ylabel('Count')
plt.xticks([0, 1], ['Group 1', 'Group 2'])
# Adjust legend if needed, e.g., handles, labels = plt.gca().get_legend_handles_labels(); plt.legend(handles, ['Sex 1', 'Sex 2'])
plt.show()
```

```python
# Calculate descriptive statistics by Group
print("\nDescriptive Statistics by Group:")
grouped_stats = train.groupby('Group')[['Age', 'PIQ', 'VIQ', 'FSIQ']].agg(['mean', 'median', 'std', 'count'])
print(grouped_stats)
```

```python
train_with_ados = train.dropna(subset=['ADOS']).copy()

if not train_with_ados.empty:
    print(f"\nAnalyzing subset with ADOS scores (N={len(train_with_ados)})")

    # Distribution of ADOS scores
    plt.figure(figsize=(8, 5))
    sns.histplot(data=train_with_ados, x='ADOS', kde=True, bins=10)
    plt.title('Distribution of ADOS Scores (where available)')
    plt.xlabel('ADOS Score')
    plt.ylabel('Frequency')
    plt.show()

    # Compare ADOS scores between Groups (if applicable)
    if train_with_ados['Group'].nunique() > 1: # Check if both groups exist in subset
        plt.figure(figsize=(7, 5))
        sns.boxplot(x='Group', y='ADOS', data=train_with_ados)
        plt.title('ADOS Score Distribution by Group (where available)')
        plt.xlabel('Group')
        plt.ylabel('ADOS Score')
        plt.xticks([0, 1], ['Group 1', 'Group 2'])
        plt.show()
    else:
        print("Not enough group diversity in the ADOS subset to compare groups.")

else:
    print("\nNo non-missing ADOS scores found in the dataset.")
```
